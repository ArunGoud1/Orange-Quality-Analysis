## Questions

2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity.
3)  How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5)).
4)  Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels.

## Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit.

```python
O =  Orange[["Size (cm)","Weight (g)"]].corr()
```

```python
O
```

```python
sns.barplot(data= O ,  x="Size (cm)",y="Weight (g)",hue = "Size (cm)") 
```

```python
sns.boxplot(data= O ,  x="Size (cm)",y="Weight (g)",hue = "Size (cm)") 
```

##  How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity.

```python
Orange.head(1)
```

```python
O1 =  Orange[["Brix (Sweetness)","pH (Acidity)"]].corr()
```

```python
O1
```

```python
sns.barplot(data=O1,x="Brix (Sweetness)",y="pH (Acidity)",hue="pH (Acidity)")
```

```python
sns.heatmap(O1,annot = True)
```

## How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))

```python
Or = Orange[["Softness (1-5)","Ripeness (1-5)"]]
```

```python
Or
```

```python
oo = Or.head(10)
```

```python
oo
```

```python
sns.barplot(data =oo ,x = "Softness (1-5)",y= "Ripeness (1-5)",hue = "Ripeness (1-5)")
```

## Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels

```python
Oo = Orange[["Blemishes (Y/N)","Ripeness (1-5)"]]
```

```python
Oo
```

```python
sns.pairplot(Orange[["Blemishes (Y/N)","Ripeness (1-5)","Softness (1-5)"]],hue = "Blemishes (Y/N)")
plt.show()
```

```python
sns.barplot(data =Oo ,x = "Blemishes (Y/N)",y= "Ripeness (1-5)",hue = "Ripeness (1-5)")
```

## How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness

```python
Oranges = Orange.groupby("Variety")["Size (cm)"].sum().reset_index()
```

```python
Oranges
```

```python
sns.barplot(data =Oranges,y = "Variety",x = "Size (cm)",hue = "Variety")
```

```python
Oranges1 = Orange.groupby("Variety")["Weight (g)"].sum().reset_index()
```

```python
Oranges1
```

```python
sns.barplot(data =Oranges1,y = "Variety",x = "Weight (g)",hue = "Weight (g)")
```

```python
Oranges2 = Orange.groupby("Variety")["Brix (Sweetness)"].sum().reset_index()
```

```python
Oranges2
```

```python
sns.barplot(data =Oranges2,y = "Variety",x = "Brix (Sweetness)",hue = "Brix (Sweetness)")
```

```python
Oranges3 = Orange.groupby("Variety")["pH (Acidity)"].sum().reset_index()
```

```python
Oranges3
```

```python
sns.barplot(data =Oranges3,y = "Variety",x = "pH (Acidity)",hue = "pH (Acidity)")
```

```python
Oranges4 = Orange.groupby("Variety")["Ripeness (1-5)"].sum().reset_index()
```

```python
Oranges4
```

```python
sns.barplot(data =Oranges4,y = "Variety",x = "Ripeness (1-5)",hue = "Ripeness (1-5)")
```
